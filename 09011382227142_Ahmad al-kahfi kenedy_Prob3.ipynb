```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import percentile
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_excel("Data Tugas Prob3_.xlsx")
#https://triatmono.info/data-penjualan-tahun-2012/data-penjualan-mobil-2017/
```

```python
sns.pairplot(df, x_vars=['Toyota', 'Mitsubishi', 'Suzuki'],
            y_vars='Total', height = 4, aspect = 1, kind = 'scatter')
plt.show()
```

```python
sns.heatmap(df.corr(), cmap = "YlGnBu", annot = True )
plt.show()
```
